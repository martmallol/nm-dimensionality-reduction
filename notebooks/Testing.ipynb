# Testeamos el cálculo de autovalores y autovectores

```python
import numpy as np
```

```python
A = np.array([
    [2, 0],
    [0, 1]
])
lambdas_exp = np.linalg.eig(A)[0]
lambdas_exp = -np.sort(-lambdas_exp) # Nos aseguramos de ordenar los autovalores de mayor a menor

lambdas, V = metnum.get_first_eigenvalues(A, 2)

print (np.allclose(lambdas_exp, lambdas))
print (np.allclose(V @ np.diag(lambdas) @ V.T, A))
```

```python
B = np.array([
    [2, 1],
    [1, 1]
])
lambdas_exp = np.linalg.eig(B)[0]
lambdas_exp = -np.sort(-lambdas_exp) # Nos aseguramos de ordenar los autovalores de mayor a menor

lambdas, V = metnum.get_first_eigenvalues(B, 2)

print (np.allclose(lambdas_exp, lambdas))
print (np.allclose(V @ np.diag(lambdas) @ V.T, B))
```

```python
C = np.array([
    [2, 0, 2, 0, 2],
    [0, 3, 0, 3, 0],
    [2, 0, 2, 0, 2],
    [0, 3, 0, 3, 0],
    [2, 0, 2, 0, 2]
])
lambdas_exp = np.linalg.eig(C)[0]
lambdas_exp = -np.sort(-lambdas_exp) # Nos aseguramos de ordenar los autovalores de mayor a menor

lambdas, V = metnum.get_first_eigenvalues(C, 5)

print (np.allclose(lambdas_exp, lambdas))
print (np.allclose(V @ np.diag(lambdas) @ V.T, C))
```

```python
# Ahora veamos que pasa si tomamos menos autovalores
alpha = 2
lambdas_new, V_new = metnum.get_first_eigenvalues(C, alpha)

print (np.allclose(lambdas[:alpha], lambdas_new))
print (np.allclose(C @ V_new, V_new @ np.diag(lambdas_new)))
```

```python
E = np.random.rand(100, 100) # Creo una matriz al azar de 100x100 con coeficientes entre 0 y 100.
E = E @ E.T # La multiplico por sí misma transpuesta para que sea simétrica.
lambdas_exp = np.linalg.eig(E)[0]
lambdas_exp = -np.sort(-lambdas_exp) # Nos aseguramos de ordenar los autovalores de mayor a menor

alpha = 30
lambdas, V = metnum.get_first_eigenvalues(E, alpha) # Veamos los primeros valores, ya que esto es lo que vamos a hacer en PCA.

print (np.allclose(lambdas_exp[:alpha], lambdas))
print (np.allclose(E @ V, V @ np.diag(lambdas)))
print (E @ V)
print ('lalalallala')
print (V @ np.diag(lambdas))
```

# Testeamos el cálculo de covarianza.

```python
alpha = 30
```

```python
pca_A = metnum.PCA(alpha)
A = np.array([
    [64, 580, 29],
    [66, 570, 33],
    [68, 590, 37],
    [69, 660, 46],
    [73, 600, 55]
])
covariance_exp = np.cov(A.T)

covariance = pca_A.covariance(A)

assert np.allclose(covariance_exp, covariance)
```

```python
pca_B = metnum.PCA(alpha)
B = np.array([
    [90, 60, 90],
    [90, 90, 30],
    [60, 60, 60],
    [60, 60, 90],
    [30, 30, 30]
])
covariance_exp = covariance_exp = np.cov(B.T)

covariance = pca_B.covariance(B)

assert np.allclose(covariance_exp, covariance)
```

```python
pca_C = metnum.PCA(alpha)
C = np.random.rand(100, 20) # Creo una matriz al azar de 100x100 con coeficientes entre 0 y 100.
covariance_exp = covariance_exp = np.cov(C.T)

covariance = pca_B.covariance(C)

assert np.allclose(covariance_exp, covariance)
```

# Testeamos kNN

```python
k = 10
A_train = np.eye(20)
y_train = np.arange(20)
A_test = A_train[:5]

knn = metnum.KNNClassifier(k)
knn.fit(A_train, y_train)
y_test = knn.predict(A_test)

assert np.allclose(y_test, y_train[:5])
```

```python
k = 3
row = np.ones(5)
A_train = np.array([
    row,
    row * 2,
    row * 5,
    row * 0.5,
    row * 50,
    row * 100
])
y_train = np.array([1, 2, 5, 0.5, 50, 100])
A_test = np.array([
    row * 3,
    row * 52,
    row * 120
])

knn = metnum.KNNClassifier(k)
knn.fit(A_train, y_train)
y_test = knn.predict(A_test)
y_exp = np.array([2, 50, 100])

assert np.allclose(y_test, y_exp)
```

#### Ahora hagamos tests de kNN usando el conjunto de dígitos que tenemos. Usemos el train para poder comparar los resultados.

```python
import pandas as pd

train_set = pd.read_csv('../data/train.csv')
display(train_set)
```

```python
train_data = train_set.sample(1000)
display(train_data)

X_train = train_data[train_data.columns[1:]].values
y_train = train_data["label"].values
```

```python
test_data = train_set.sample(10)
display(test_data)

X_test = test_data[test_data.columns[1:]].values
y_exp = test_data["label"].values
```

```python
k = 10
knn = metnum.KNNClassifier(k)
knn.fit(X_train, y_train)
y_test = knn.predict(X_test)

ratio = 0
for i in range(len(y_test)):
    if y_test[i] == y_exp[i]:
        ratio += 1
ratio = ratio / len(y_test)
print(ratio)
```

#### Si el algoritmo estuviera mal implementado y los valores de y_test fueran al azar entonces esperaríamos por probabilidades que el ratio sea de aproximadamente 0.1 (cada dígito es una variable aleatoria con distribución \\( Be(0.1) \\) así que la suma es una variable aleatoria con distribución \\( B(10, 0.1) \\)). Sin embargo, si corremos las últimas tres celdas varias veces la mayoría de las veces el ratio es mayor a 0.1, así que el algoritmo está funcionando. Además, si incrementamos el tamaño del conjunto de entrenamiento entonces vamos a notar que el ratio se acerca a 1.

# Testeamos PCA

```python
train_data = train_set.sample(1000)
display(train_data)

X_train = train_data[train_data.columns[1:]].values
y_train = train_data["label"].values
```

```python
test_data = train_set.sample(10)
display(test_data)

X_test = test_data[test_data.columns[1:]].values
y_exp = test_data["label"].values
```

```python
alpha = 30
pca = metnum.PCA(alpha)
pca.fit(X_train)
X_train_trans = pca.transform(X_train)
X_test_trans = pca.transform(X_test)
```

```python
k = 10
knn = metnum.KNNClassifier(k)
knn.fit(X_train_trans, y_train)
y_test = knn.predict(X_test_trans)

ratio = 0
for i in range(len(y_test)):
    if y_test[i] == y_exp[i]:
        ratio += 1
ratio = ratio / len(y_test)
print(ratio)
```

#### Acá vemos lo mismo que vimos para kNN. Por lo tanto, confiamos en la implementación de PCA.
